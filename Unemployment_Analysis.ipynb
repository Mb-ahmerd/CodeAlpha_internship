```python
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=df)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Estimated Unemployment Rate')
sns.lineplot()
```

```python
plt.figure(figsize=(2,4))
sns.barplot(x='Area', y='Estimated Unemployment Rate (%)', data=df)
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(x='Area', y='Estimated Unemployment Rate (%)', data=df)
```

```python
region_unemployment_df = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).reset_index()
region_unemployment_df.columns = ['Region', 'Average Unemployment Rate']

plt.figure(figsize=(10, 8))
sns.barplot(y='Region', x='Average Unemployment Rate', data=region_unemployment_df)
plt.title('Average Unemployment Rate by Region')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Region')
plt.show()
```

```python
region_area_unemployment_over_time = df.groupby(['Date', 'Area'])['Estimated Unemployment Rate (%)'].mean().reset_index()

plt.figure(figsize=(12, 10))
sns.lineplot(
    y='Estimated Unemployment Rate (%)',
    x='Date',
    hue='Area',
    data=region_area_unemployment_over_time,
    marker='o' )
plt.title('Average Unemployment Rate (Rural vs. Urban) Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(title='Area')
plt.tight_layout()
plt.show()
```

## Overview
This analysis explores unemployment trends in India using monthly data across 28 states.
The dataset covers the period from May 2019 to June 2020, capturing the significant
impact of the COVID-19 pandemic on employment.

## Key Findings

### 1. Covid-19 Impact
The COVID-19 pandemic caused a sharp increase in unemployment in India, with the highest rates recorded during April and May 2020 due to nationwide lockdowns. Although restrictions were gradually eased from July onward, unemployment remained high before beginning to decline as economic activities resumed.

### 2. Regional Unemployment Analysis
Unemployment rates varied significantly across Indian states, with Tripura recording the highest rate and Meghalaya the lowest. Overall, northern states experienced higher unemployment than many southern states, reflecting regional differences in economic activity and employment opportunities.
### 3. Rural vs Urban Comparison
Urban areas consistently had higher unemployment rates than rural areas, particularly during the COVID-19 lockdown when industries and businesses were forced to close. Rural areas were less affected because many residents relied on agriculture and small-scale businesses that continued operating.

### 4. Feature Correlations
The correlation analysis shows a weak negative relationship (-0.22) between unemployment rate and estimated employment, indicating that unemployment generally decreases as employment increases. Labour participation rate has almost no relationship with unemployment, suggesting that workforce participation is largely independent of unemployment levels in this dataset.


## Conclusion
The analysis shows that COVID-19 significantly increased unemployment across India, especially in urban areas and certain northern states. These findings highlight the influence of regional economic differences and demonstrate how exploratory data analysis can reveal important patterns for further predictive modeling.
